```python
def show(image):
    im = Image.open(image)
    im.show()
```

```python
# From - https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# Loading image
file = datadir + i + '.jpg'
img = cv.imread(file)
print (img)
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```
